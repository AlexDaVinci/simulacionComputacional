```python
#Generador MidSquare
def midSquare(seed, n, N):
  ciclo = None
  cicloEncontrado = False
  numerosAleatorios = []
  for i in range(N):
    seed = str(seed)
    if len(seed) < 2 * n:
      seed = '0' * (2 * n - len(seed)) + seed
    seed = int(seed)
    seedCuadrado = seed ** 2
    seedCuadrado = str(seedCuadrado)
    longitudSeedCuadrado = len(seedCuadrado)
    inicio = (longitudSeedCuadrado // 2) - n
    fin = inicio + 2 * n
    nuevoNumero = int(seedCuadrado[inicio:fin])
    numeroDecimal = nuevoNumero / (10 ** (2 * n))
    numerosAleatorios.append(numeroDecimal)
    if numeroDecimal in numerosAleatorios[:i] and cicloEncontrado == False:
      ciclo = i - numerosAleatorios.index(numeroDecimal)
      cicloEncontrado = True
    seed = nuevoNumero
  return numerosAleatorios, ciclo
```

```python
#Generador congruencialMixto
def congruencialMixto(x0, a, c, m, N):
  ciclo = None
  cicloEncontrado = False
  xn = x0
  numeroDecimal = xn / (10 ** len(str(x0)))
  numerosGenerados = [numeroDecimal]
  for i in range(N):
    xn = (a * xn + c) % m
    numeroDecimal = xn / (10 ** len(str(x0)))
    numerosGenerados.append(numeroDecimal)
    if numeroDecimal in numerosGenerados[:i] and cicloEncontrado == False:
      ciclo = i
      cicloEncontrado = True
  return numerosGenerados, ciclo
```

Para la secuencia de $N=1000$ números pseudo-aleatorios de cada uno de los siguientes generadores aleatorios realice el test de uniformidad Chi-cuadrado ($k=100$) y el test de aleatoriedad de rachas, calculando el respectivo estadístico respectivo y verificando la hipótesis nula o la hipótesis de rechazo, para confirmar si la secuencia cumple con o no con los criterios de uniformidad y aleatoriedad. Use un 95% de intervalo de confianza.




```python
#Generador congruencialMixto
def generar_numero(x_anterior, a, c, m):
    x_siguiente = (a * x_anterior + c) % m
    u_siguiente = x_siguiente / m
    return [x_siguiente, u_siguiente]

def congruencialMixto(x0, a, c, m, N):
    x_anterior = x0
    numerosGenerados = []

    for t in range(1, N + 1):
        xi, ui = generar_numero(x_anterior, a, c, m)
        numerosGenerados.append(ui)
        x_anterior = xi

    return numerosGenerados
```

```python
def prueba_chi_cuadrado(U, k, alpha):
    N = len(U)
    frec_obs, bin_edges = np.histogram(U, bins=k)
    frec_esp = N / k
    chi2_valor=0
    for i in range(k):
      chi2_valor= chi2_valor + ((frec_obs[i] - frec_esp)**2) / frec_esp

    print("Chi2 = ", chi2_valor)

    chi2_critico = stats.chi2.ppf(1 - alpha, k - 1)
    print("Chi2_k-1 alpha = ", chi2_critico)

    if chi2_valor > chi2_critico:
        print("Se rechaza la hipótesis de uniformidad.")
    else:
        print("Se acepta la hipótesis de uniformidad.")
```

```python
def prueba_rachas(U, alpha):
    N = len(U)
    arreglo = np.zeros(N - 1)

    for i in range(1, N):
        if U[i - 1] > U[i]:
            arreglo[i - 1] = 0
        else:
            arreglo[i - 1] = 1

    boolArreglo = np.array((arreglo == 1))
    signo = (boolArreglo[:-1] ^ boolArreglo[1:])
    isigno = np.arange(len(signo))[signo]
    isigno = list(isigno)
    isigno.insert(0, -1)
    isigno = np.array(isigno)
    L = isigno[1:(len(isigno))] - isigno[0:(len(isigno) - 1)]
    R = len(L)

    print("R = ", R)

    Z = (R - ((2 * N - 1) / 3)) / (np.sqrt((16 * N - 29) / 30))
    print("Z = ", Z)

    Zt = stats.norm.ppf((1 + (1 - alpha)) / 2)
    print("Z_alpha/2 = ", Zt)

    if abs(Z) > Zt:
        print("Se rechaza la hipótesis de aleatoriedad.")
    else:
        print("Se acepta la hipótesis de aleatoriedad.")
```
